```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#Visualization settings
sns.set_style(style='white')
sns.set(rc={
    'figure.figsize': (12,7),
    'axes.facecolor': 'white',
    'axes.grid': True,
    'grid.color': '.9',
    'axes.linewidth' : 1.0,
    'grid.linestyle': u'-'},
    font_scale=1.5)
custom_colors = ["#3498db", "#95a5a6","#34495e", "#2ecc71", "#e74c3c"]
sns.set_palette(custom_colors)
```

## <center> relplot </center>

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(data=tips,x="total_bill",y="tip", hue="day", col="time", row="sex");
```

# <center> 1.Scatterplot </center>

```python
sns.scatterplot(data=tips,x="total_bill", y="tip");
```

# <center>2.Lineplot </center>

```python
flights = sns.load_dataset("flights")
flights
```

```python
Jan_flights = flights.query("month == 'January'" )
sns.lineplot(data=Jan_flights, x="year", y="passengers");
```

```python
sns.lineplot(data=flights, x="year", y="passengers");
print ("Passing the entire dataset in long-form mode will aggregate over repeated values (each year) to show the mean and 95% confidence interval:")
```

```python
sns.lineplot(data=flights,x="year", y="passengers", hue="month");
print ("Assign a grouping semantic (hue, size, or style) to plot separate lines")
```

# <center>3.Catplot </center>
Shows the relationship between a numerical variable and one more categorical variables.

```python
exercise = sns.load_dataset("exercise")
```

```python
exercise
```

```python
sns.catplot(data=exercise, x="time", y="pulse", hue="kind");
```

# <center>4.Stripplot</center>
Draw a scatterplot where one variable is categorical.

```python
sns.stripplot(data=tips,x="total_bill",linewidth=1);
```

```python
sns.stripplot(data=tips, x="day",y="total_bill",linewidth=1);
```

# <center>5.Swarmplot</center>

```python
sns.swarmplot(x="total_bill", data=tips);
```

```python
sns.swarmplot(data=tips,x="total_bill",y="day");
```

```python
sns.swarmplot(data=tips,x="day",y="total_bill",hue="smoker",dodge=True);
```

# <center>6. Boxplot </center>
Draw a box plot to show distributions with respect to categories.

```python
sns.boxplot(data=tips,x="total_bill");
```

```python
sns.boxplot(data=tips,y="total_bill",x="day");
```

# <center> 12. Distplot</center>
This function combines the matplotlib hist function (with automatic calculation of a good default bin size) with the seaborn kdeplot() and rugplot() functions. It can also fit scipy.stats distributions and plot the estimated PDF over the data.

```python
x=np.random.randn(100)
sns.distplot(x);
```

# <center> 13. Kdeplot</center>

```python
sns.kdeplot(data=tips.total_bill);
```

# <center> 14. Rugplot</center>
Plot marginal distributions by drawing ticks along the x and y axes.

This function is intended to complement other plots by showing the location of individual observations in an unobstrusive way.

```python
sns.kdeplot(data=tips.total_bill);
sns.rugplot(a=tips.total_bill);
```

# <center>15. lmplot </center>
Plot data and regression model fits across a FacetGrid.

```python
sns.lmplot(data=tips,x="total_bill", y="tip");
```

# <center> 16.regplot </center>
Plot data and a linear regression model fit.

```python
sns.regplot(data=tips,x= 'total_bill', y='tip');
```
