# <center>Multiple plots in same graph</center>

```python
basket=[]
apple=("Apple," * random.randint(1,100)).split(",")
grapes=("Grape," * random.randint(1,100)).split(",")
orange=("Orange," * random.randint(1,100)).split(",")
basket.extend(apple)
basket.extend(grapes)
basket.extend(orange)
basket=pd.Series(data=basket,name="Item")
basket.drop(basket[basket.values==""].index,inplace=True)
basket_value_cnt=basket.value_counts()
cum_percent=basket_value_cnt.cumsum()/basket_value_cnt.sum()*100
```

```python
sns.reset_defaults()
```

```python
fig,ax=plt.subplots(figsize=(20,10))
ax.bar(basket_value_cnt.index,basket_value_cnt.values,color='lightblue');

ax2=ax.twinx()
ax2.plot(cum_percent.index,cum_percent,color='red',marker='D',ms=5)
ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis='y', colors='blue')
ax2.tick_params(axis='y',colors='red')
plt.show()
```

# <center>Plot customizations</center>

```python
sns.reset_defaults()
fig=plt.figure(figsize=(22,15));
gs=fig.add_gridspec(3,3)
gs.update(wspace=0.35,hspace=0.27)

#Create subplots
ax0=fig.add_subplot(gs[0,0]);
ax1=fig.add_subplot(gs[0,1]);
ax2=fig.add_subplot(gs[1,0]);
ax3=fig.add_subplot(gs[1,1]);



#Setting the background color for entire figure as well as graph plots
background_color="#fbfbfb"
fig.patch.set_facecolor(background_color)
ax0.set_facecolor(background_color)
ax1.set_facecolor(background_color)
ax2.set_facecolor(background_color)
ax3.set_facecolor(background_color)

#Plugging actual graphs for each subplots
sns.scatterplot(data=tips,x="total_bill", y="tip",ax=ax0).set_title("graph1");
sns.scatterplot(data=exercise, x="time", y="pulse",ax=ax1).set_title("graph2");

#Displaying text in axis place
ax3.spines["bottom"].set_visible(False)
ax3.spines["left"].set_visible(False)
ax3.tick_params(left=False,bottom=False)
ax3.set_xticklabels([])
ax3.set_yticklabels([])

ax3.text(x=0.5,
          y=0.5,
          s="Some Text\nby hiding all margins\nin this axis",
          horizontalalignment='center',
          verticalalignment='center',
          fontsize=22,
          fontweight='bold',
          fontfamily='serif',
          color='grey'
         )

for position in ["top","right"]:
    ax0.spines[position].set_visible(False)
    ax1.spines[position].set_visible(False)
    ax2.spines[position].set_visible(False)
    ax3.spines[position].set_visible(False)
    
    
l1=lines.Line2D([0.65,0.65],[0.299,0.9], transform=fig.transFigure, figure=fig, color='black',lw=0.2)
fig.lines.extend([l1])

fig.text(x=0.75,y=.8,s="Conclusion",
        fontsize=20,
        fontweight='bold',
        fontfamily='serif',
        color='grey')

fig.text(x=0.66,y=0.39,s='''
Below is the sample text\n
We note that most job-seekers are Male.
This is not all that surprising as in this
dataset Males make up the majority of the
sample population.

What is more interesting though is
the City Development Index (CDI) chart.
There we see that there are two peaks
for job-seekers. 
The peaks are at high and low CDI scores. 


We can ponder why this might be;
in high CDI areas perhaps there are a 
lot of opportunities and therefore
people feel encouraged to seek better roles.
''',
        fontsize=20,
        fontweight='light',
        fontfamily='serif',
        color='grey')
plt.show()
```
